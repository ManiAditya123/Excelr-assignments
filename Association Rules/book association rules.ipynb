```python
a=apriori(data, use_colnames=True, min_support=0.1)
a
```

```python
b=association_rules(a, min_threshold=0.7, metric='lift')
b
```

```python
b.sort_values('lift', ascending=False)
```

```python
b[b.lift>1]
```

```python
import matplotlib.pyplot as plt
plt.scatter(b['support'],b['confidence'])
plt.xlabel('support')
plt.ylabel('confidence')
plt.show()
```
